```python
import pandas as pd
```

```python
df = pd.read_excel('dividend_aristocrats.xlsx')
```

```python
df['quintile'] = pd.qcut(df['Dividend Yield'], 5, labels=False)
```

```python
df
```

```python
df_grouped = df.groupby('quintile')
```

```python
df_grouped.groups
```

```python
df_target = dict(list(df_grouped))[3]
```

```python
df_target
```

```python
df_target_sorted = df_target.sort_values('Dividend Yield', ascending=False)
```

```python
df_target_sorted
```

```python
df_target_sorted.to_csv('Dividend Aristocrats 2nd group of Quintile.csv')
```
