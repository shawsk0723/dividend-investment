```python
import pandas_datareader as pdr
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
#datetime으로 시작과 종료 날짜를 만든다.
start = datetime.datetime (1900, 1, 1)
end = datetime.datetime (2023, 5, 4)
```

```python
#datareader를 사용하여 FRED에서 데이터를 가져온다.
df = pdr.DataReader('FEDFUNDS', 'fred', start, end)
df.head(3)
```

```python
df.tail(3)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
#데이터 시각화하기
plt.plot(df.index, df.FEDFUNDS)
plt.xlabel('년도')
plt.ylabel('금리')
plt.title('연준 금리')
plt.show()
```

```python
#데이터 시각화하기
plt.plot(df.index, df.FEDFUNDS.diff(1))
plt.xlabel('년도')
plt.ylabel('금리')
plt.title('연준 금리')
plt.show()
```

```python
df_after_2000 = df[df.index >= '2021-01-01']
df_after_2000.head(3)
```

```python
#데이터 시각화하기
plt.plot(df_after_2000.index, df_after_2000.FEDFUNDS)
plt.grid()
plt.xlabel('년도')
plt.ylabel('금리')
plt.title('연준 금리')
plt.xticks(rotation=45)
plt.show()
```

```python
#데이터 시각화하기
plt.plot(df_after_2000.index, df_after_2000.FEDFUNDS.diff())
plt.xlabel('년도')
plt.ylabel('금리')
plt.title('연준 금리 변화량')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#데이터 시각화하기


fig, ax = plt.subplots()
ax.plot(df_after_2000.index, df_after_2000.FEDFUNDS, label='금리')
ax.grid()
ax.set_xlabel('년도')
ax.set_ylabel('금리')
ax.legend(loc='upper left')
ax.tick_params(axis='x', rotation=45)

ax2 = plt.twinx()
ax2.plot(df_after_2000.index, df_after_2000.FEDFUNDS.diff(), 
         label='금리 변화량',
         linestyle='--',
         color='r')
ax2.set_ylabel('금리 변화량')
ax2.legend(loc='lower right')

plt.title('연준 금리')
plt.show()
```

```python
#데이터 시각화하기
plt.plot(df_after_2000.index, df_after_2000.FEDFUNDS.diff().diff())
plt.xlabel('년도')
plt.ylabel('금리')
plt.title('연준 금리 변화량')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
